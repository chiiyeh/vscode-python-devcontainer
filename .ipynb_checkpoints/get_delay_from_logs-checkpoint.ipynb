```python
import csv
from datetime import datetime
import re
```

```python
# CA1b2a38d2901a97d69701860abf111295 IN ECDA call
with open('CA7ce96e4b012729d54b7842d9e7aec6c6.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    header = True
    starttime = ''
    prev_line = ''
    for row in csv_reader:
        if header:
            header = False
            continue
        if not starttime:
            starttime = datetime.strptime(row[1], '%Y-%m-%d %H:%M:%S.%f')
            continue
        if re.search("INFO", row[2]):
            if match:=re.search("'total-length': (\d+\.\d+)", row[2]):
                time_diff = datetime.strptime(row[1], '%Y-%m-%d %H:%M:%S.%f') - starttime
                print(time_diff.total_seconds()-float(match.group(1)))
                print(prev_line)
            else:
                prev_line = row[2]
```

```python
# CA1b2a38d2901a97d69701860abf111295 IN ECDA call
with open('log/2add2a86-2dc7-4ed3-9b5e-b5b021ff5049 - worker.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    header = True
    starttime = ''
    prev_line = ''
    for row in csv_reader:
        if re.search("Initializing request", row[2]):
            starttime = datetime.strptime(row[1], '%Y-%m-%d %H:%M:%S.%f')
            print(f"Starttime: {starttime}")
            continue
        if re.search("After postprocessing", row[2]) and starttime:
            if match:=re.search("'total-length': (\d+\.\d+)", row[2]):
                current_time = datetime.strptime(row[1], '%Y-%m-%d %H:%M:%S.%f')
                time_diff = current_time - starttime
                print(f"Current time: {current_time}")
                print(time_diff.total_seconds()-float(match.group(1)))
                print(row[2])
```
